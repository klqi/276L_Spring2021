```python
import pandas as pd 
import numpy as np
```

```python
all_data = pd.read_csv('habsos_20200310.csv')
# remove all not-observed stations
filtered = all_data[all_data['category'] != 'not observed']
filtered.sample_date = pd.to_datetime(filtered.sample_date)
#filtered.to_csv('hab_filtered_2020.csv')
filtered = filtered.groupby(pd.DatetimeIndex(filtered.sample_date).to_period('D')).nth(0)
filtered['month'] = pd.DatetimeIndex(filtered.sample_date).to_period('M')
```
